# Problem statement-Given MaxTemp is an independant feature, we have to predict min temperature 

```python
#Independant feature and dependant feature
x=df1['MaxTemp'].values.reshape(-1,1)  #since we have only 1 feature, reshape into an array
y=df1['MinTemp'].values.reshape(-1,1)
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
#object class
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr_pred=lr.predict(X_test)
```

```python
lr_pred
```

```python
from sklearn.metrics import r2_score
```

```python
#Accuracy
r2_score(y_test,lr_pred)*100  
```

```python
#accuracy on test data
lr.score(X_test,y_test)*100
```

```python
#accuracy on train data
lr.score(X_train,y_train)*100
```

# So model is not overfitted since test and train data accuracy is almost same
